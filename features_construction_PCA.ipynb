```python
import pandas as pd
```

```python
feats = pd.read_csv('features/Median_HARMrep_vowels.csv', delimiter=',')
feats = feats.set_index('SoundFile')
feats.index.names=['names']
feats.index = feats.index.str.lower()
feats.head()
```

```python
print(len(feats.index))
```

```python
formants = pd.read_csv('features/FORMANTS.csv')
formants = formants.set_index('vowels')
formants.index.names=['names']
formants.index = formants.index.str.lower()
formants.index = formants.index.str.replace('keti grei', 'keti_grei')
formants.head()
```

```python
print(len(formants.index))
```
