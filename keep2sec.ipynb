```python
import os
import numpy as np
import shutil
```

```python
os.makedirs('Rebetika_2sec', exist_ok=True)
os.makedirs('Rebetika_2sec/train', exist_ok=True)
os.makedirs('Rebetika_2sec/test', exist_ok=True)
```

```python
main_folder = 'Rebetika_2sec'
for train_test in ['train', 'test']:
    os.makedirs( main_folder + os.sep + train_test, exist_ok=True )
    print(train_test)
    for singer in os.listdir( 'Rebetika_2sec_full' + os.sep + train_test ):
        print(singer)
        os.makedirs( main_folder + os.sep + train_test + os.sep + singer, exist_ok=True )
        # keep list of all 2sec recordings
        list2sec = os.listdir( 'Rebetika_2sec_full' + os.sep + train_test + os.sep + singer )
        # make a random permutation of indexes of files
        idxs = np.random.permutation( len( list2sec ) )
        # keep the proper number of indexes, equal to number of available vowels
        num_vowels = len( os.listdir('Rebetika_vowels' + os.sep + train_test + os.sep + singer) )
        print('num_vowels: ', num_vowels)
        idxs2keep = idxs[:num_vowels]
        for idx in idxs2keep:
            src = 'Rebetika_2sec_full' + os.sep + train_test + os.sep + singer + os.sep + list2sec[idx]
            dst = main_folder + os.sep + train_test + os.sep + singer + os.sep + list2sec[idx]
            shutil.copyfile( src, dst )
```
